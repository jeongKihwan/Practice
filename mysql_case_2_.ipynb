```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('http://corners.gmarket.co.kr/BestSellers')
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
categories = soup.select('#categoryTabG > li > a')
categories
```

```python
base_url = 'http://corners.gmarket.co.kr'
```

```python
# main category, sub category 크롤링

for category in categories:
#     print(category)
    cate_title = category.text
    cate_url = base_url + category['href']
    
    res = requests.get(cate_url)
    soup = BeautifulSoup(res.content, 'html.parser')
    
    sub_categories = soup.select('div.cate-l > div.navi.group > ul > li > a')
    
    for sub_category in sub_categories:
        sub_cate_name = sub_category.text
        sub_cate_url = base_url + sub_category['href']
    
        print(cate_title, cate_url, sub_cate_name, sub_cate_url)
    
```

```python
# 상품정보를 크롤링하는 코드..

url = 'http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G01&subGroupCode=S162'

res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')

lis = soup.select('div.best-list > ul > li')

for li in lis[5:]:
    prod_ranking = li.select('p')[0].text
    prod_title = li.select('div.thumb > a > img')[0]['alt']
    prod_url = li.select('div.thumb > a')[0]['href']
    old_price = li.select('div.item_price > div.o-price')[0].text.replace(',','')[:-1]
    dis_price = li.select('div.item_price > div.s-price')[0].text.split('원')[0].replace(',','')[:-1]
    dis_percent = li.select('div.item_price > div.s-price')[0].text.split('원')[1].strip()[:-1]
    
    res = requests.get(prod_url)
    soup = BeautifulSoup(res.content, 'html.parser')
    
    provider = soup.select('div.item-topinfo > div.item-topinfo_headline > p > a > strong')[0].text
    item_code = soup.select('div.item-topinfo > div.item-topinfo_headline > p > span > span')[0].text.split(':')[1].strip()


    print(prod_ranking, prod_title, prod_url, old_price, dis_price, dis_percent, provider, item_code)
    print('='*30)
```
